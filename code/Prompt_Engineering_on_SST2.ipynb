```python
# Setup: Install dependencies
!pip install -q transformers datasets accelerate bitsandbytes peft
!pip install -U datasets huggingface_hub

# Imports
from datasets import load_dataset
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
import re
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import f1_score
from tqdm import tqdm
from collections import Counter

# === [1] Model Setup ===
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    quantization_config=bnb_config,
    torch_dtype=torch.float16
)

text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=20,
    temperature=0.3,
    pad_token_id=tokenizer.eos_token_id
)

# === [2] Dataset Loading ===
ds = load_dataset("stanfordnlp/sst2")
dataset = ds["validation"]
label_map = {0: "negative", 1: "positive"}

# === [3] Prompt Templates ===
def zero_shot_prompt(sentence):
    return f"Classify the sentiment of the following sentence:\n\"{sentence.strip()}\"\nSentiment:"

def few_shot_prompt(sentence):
    examples = [
        ("I really loved the movie, it was fantastic!", "positive"),
        ("The film was boring and too long.", "negative"),
        ("An excellent performance by the lead actor.", "positive"),
        ("I did not enjoy the movie at all.", "negative")
    ]
    context = "\n".join([f"Sentence: \"{ex[0]}\"\nSentiment: {ex[1]}" for ex in examples])
    return f"Classify the sentiment of the following sentence. Choose from: positive, negative.\n{context}\nSentence: \"{sentence.strip()}\"\nSentiment:"

# === [4] Evaluation Function ===
def evaluate_sst2(dataset, prompt_fn, name):
    batch_size = 16
    model_answers, true_answers, correct_flags, solved_flags = [], [], [], []

    for i in tqdm(range(0, len(dataset), batch_size), desc=f"Evaluating {name}"):
        batch = dataset[i:i+batch_size]
        prompts = [prompt_fn(ex["sentence"]) for ex in batch]
        golds = [label_map[ex["label"]] for ex in batch]

        outputs = text_generator(prompts)

        for j, output in enumerate(outputs):
            full_text = output[0]['generated_text']
            pred_text = full_text.replace(prompts[j], "").strip().lower()

            if "positive" in pred_text:
                pred = "positive"
            elif "negative" in pred_text:
                pred = "negative"
            else:
                pred = "unknown"

            correct = pred == golds[j]
            solved = pred in ["positive", "negative"]
            model_answers.append(pred)
            true_answers.append(golds[j])
            correct_flags.append(correct)
            solved_flags.append(solved)

    accuracy = sum(correct_flags) / len(correct_flags) * 100
    solve_rate = sum(solved_flags) / len(solved_flags) * 100
    y_true = [1 if x == "positive" else 0 for x in true_answers]
    y_pred = [1 if x == "positive" else 0 for x in model_answers]
    f1 = f1_score(y_true, y_pred)

    print(f"\n{name.title()} Accuracy: {accuracy:.2f}%")
    print(f"F1 Score: {f1:.2f}")
    print(f"Solve Rate: {solve_rate:.2f}%")

    return {
        "accuracy": accuracy,
        "f1": f1 * 100,
        "solve_rate": solve_rate,
        "model_answers": model_answers,
        "true_answers": true_answers,
        "correct_flags": correct_flags
    }

# === [5] Visualization Functions ===
def plot_summary_bar(results):
    techniques = list(results.keys())
    accuracy = [results[t]["accuracy"] for t in techniques]
    f1 = [results[t]["f1"] for t in techniques]
    solve = [results[t]["solve_rate"] for t in techniques]

    x = np.arange(len(techniques))
    width = 0.25
    plt.figure(figsize=(10, 6))
    plt.bar(x - width, accuracy, width, label="Accuracy")
    plt.bar(x, f1, width, label="F1 Score")
    plt.bar(x + width, solve, width, label="Solve Rate")
    plt.xticks(x, [t.replace("_", " ").title() for t in techniques])
    plt.ylabel("Percentage")
    plt.title("SST-2 Prompting Evaluation Summary")
    plt.ylim(0, 100)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

def plot_prediction_distribution(preds, name):
    count = Counter(preds)
    plt.figure(figsize=(6, 5))
    plt.bar(count.keys(), count.values(), color='steelblue')
    plt.title(f"Prediction Distribution - {name.title()}")
    plt.ylabel("Count")
    plt.grid(axis="y")
    plt.tight_layout()
    plt.show()

def plot_failure_pie(true_answers, preds, name):
    categories = {"Correct": 0, "Incorrect (Wrong Class)": 0, "Unknown": 0}
    for true, pred in zip(true_answers, preds):
        if pred == true:
            categories["Correct"] += 1
        elif pred in ["positive", "negative"]:
            categories["Incorrect (Wrong Class)"] += 1
        else:
            categories["Unknown"] += 1

    plt.figure(figsize=(6, 6))
    plt.pie(categories.values(), labels=categories.keys(), autopct='%1.1f%%')
    plt.title(f"Failure Analysis - {name.title()}")
    plt.tight_layout()
    plt.show()

# === [6] Run and Visualize ===
results = {
    "zero_shot": evaluate_sst2(dataset, zero_shot_prompt, "zero_shot"),
    "few_shot": evaluate_sst2(dataset, few_shot_prompt, "few_shot")
}

plot_summary_bar(results)
for tech in results:
    plot_prediction_distribution(results[tech]["model_answers"], tech)
    plot_failure_pie(results[tech]["true_answers"], results[tech]["model_answers"], tech)
```
