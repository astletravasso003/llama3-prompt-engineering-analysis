```python
# === [1] Install Dependencies ===
!pip install -q transformers datasets accelerate bitsandbytes peft
!pip install --upgrade datasets transformers

# === [2] Imports ===
from datasets import load_dataset
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from sklearn.metrics import f1_score
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import torch
from tqdm import tqdm
import re

# === [3] Model Setup (Meta LLaMA-3 8B Instruct) ===
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    quantization_config=bnb_config,
    torch_dtype=torch.float16
)

text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=30,
    temperature=0.3,
    pad_token_id=tokenizer.eos_token_id
)

# === [4] Load CommonsenseQA Dataset ===
ds = load_dataset("tau/commonsense_qa")
dataset = ds["validation"]  # Evaluation on validation split

# === [5] Prompt Templates ===
def zero_shot_prompt(q, choices):
    options = "\n".join([f"{label}. {text}" for label, text in zip(choices["label"], choices["text"])])
    return (
        "You are a knowledgeable assistant answering multiple choice questions based on commonsense reasoning.\n\n"
        f"Question: {q}\nChoices:\n{options}\nAnswer:"
    )

def few_shot_prompt(q, choices):
    examples = [
        {"question": "Where do people usually go to buy medicine?", "choices": ["pharmacy", "library", "bank", "restaurant", "school"], "answer": "A"},
        {"question": "What do people use to cut paper?", "choices": ["scissors", "fork", "pencil", "phone", "chair"], "answer": "A"},
        {"question": "Where can you find books to read?", "choices": ["park", "library", "mall", "restaurant", "bank"], "answer": "B"}
    ]
    context = ""
    for ex in examples:
        opts = "\n".join([f"{chr(65+i)}. {opt}" for i, opt in enumerate(ex["choices"])])
        context += f"Q: {ex['question']}\nChoices:\n{opts}\nAnswer: {ex['answer']}\n\n"
    opts = "\n".join([f"{label}. {text}" for label, text in zip(choices["label"], choices["text"])])
    return (
        "You are a knowledgeable assistant answering multiple choice questions based on commonsense reasoning.\n\n"
        f"{context}Q: {q}\nChoices:\n{opts}\nAnswer:"
    )

def cot_prompt(q, choices):
    opts = "\n".join([f"{label}. {text}" for label, text in zip(choices["label"], choices["text"])])
    return (
        "You are a reasoning assistant solving commonsense problems step-by-step.\n\n"
        f"Q: {q}\nChoices:\n{opts}\n\nLet's think step by step.\nAnswer:"
    )

# === [6] Evaluation Function ===
def evaluate_commonsenseqa(prompt_type):
    correct_flags = []
    model_answers = []
    true_answers = []

    for ex in tqdm(dataset, desc=f"Evaluating {prompt_type}"):
        q, choices, gold = ex["question"], ex["choices"], ex["answerKey"]

        if prompt_type == "zero_shot":
            prompt = zero_shot_prompt(q, choices)
        elif prompt_type == "few_shot":
            prompt = few_shot_prompt(q, choices)
        elif prompt_type == "cot":
            prompt = cot_prompt(q, choices)
        else:
            raise ValueError("Unknown prompt type")

        output = text_generator(prompt)[0]["generated_text"]
        response = output.split("Answer:")[-1].strip().upper()

        pred = "unknown"
        for label in ["A", "B", "C", "D", "E"]:
            if response.startswith(label):
                pred = label
                break

        model_answers.append(pred)
        true_answers.append(gold)
        correct_flags.append(pred == gold)

    accuracy = sum(correct_flags) / len(correct_flags) * 100
    solve_rate = sum([p in ["A", "B", "C", "D", "E"] for p in model_answers]) / len(model_answers) * 100

    label_to_idx = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}
    y_true = [label_to_idx.get(label, -1) for label in true_answers]
    y_pred = [label_to_idx.get(label, -1) for label in model_answers]
    f1 = f1_score(y_true, y_pred, average="macro")

    print(f"\n{prompt_type.title()} Accuracy: {accuracy:.2f}% | F1 Score: {f1 * 100:.2f}% | Solve Rate: {solve_rate:.2f}%")
    return {
        "accuracy": accuracy,
        "f1": f1 * 100,
        "solve_rate": solve_rate,
        "correct_flags": correct_flags,
        "model_answers": model_answers,
        "true_answers": true_answers
    }

# === [7] Visualization ===
def plot_summary_bar(results):
    techniques = list(results.keys())
    accuracy = [results[t]["accuracy"] for t in techniques]
    f1 = [results[t]["f1"] for t in techniques]
    solve = [results[t]["solve_rate"] for t in techniques]

    x = np.arange(len(techniques))
    width = 0.25
    plt.figure(figsize=(10, 6))
    plt.bar(x - width, accuracy, width, label="Accuracy")
    plt.bar(x, f1, width, label="F1 Score")
    plt.bar(x + width, solve, width, label="Solve Rate")
    plt.xticks(x, [t.replace("_", " ").title() for t in techniques])
    plt.ylabel("Percentage")
    plt.title("CommonsenseQA - Prompt Evaluation Summary")
    plt.ylim(0, 100)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

def plot_prediction_distribution(preds, name):
    count = Counter(preds)
    plt.figure(figsize=(6, 5))
    plt.bar(count.keys(), count.values())
    plt.title(f"Prediction Distribution - {name.title()}")
    plt.ylabel("Count")
    plt.grid(axis="y")
    plt.tight_layout()
    plt.show()

def plot_failure_pie(true_answers, model_answers, name):
    labels = ["Correct", "Wrong Option", "Format Error"]
    values = [0, 0, 0]
    for t, p in zip(true_answers, model_answers):
        if p == t:
            values[0] += 1
        elif p in ["A", "B", "C", "D", "E"]:
            values[1] += 1
        else:
            values[2] += 1
    plt.figure(figsize=(6, 6))
    plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=140)
    plt.title(f"Failure Analysis - {name.title()}")
    plt.tight_layout()
    plt.show()

# === [8] Run All Evaluations ===
results = {}
for tech in ["zero_shot", "few_shot", "cot"]:
    results[tech] = evaluate_commonsenseqa(tech)

# === [9] Visualize Results ===
plot_summary_bar(results)
for name in results:
    plot_prediction_distribution(results[name]["model_answers"], name)
    plot_failure_pie(results[name]["true_answers"], results[name]["model_answers"], name)
```
